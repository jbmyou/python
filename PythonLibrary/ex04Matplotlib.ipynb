```python
import matplotlib.pyplot as plt
```

```python
# line chart그리기
# 데이터형식 : list
x = [1,2,3]
y = [4,6,8]
plt.plot(x,y,ls='-', marker='v', lw=3, c='y', markerfacecolor='red')
#속성들
#1. ls 라인 스타일.  ='문자'
#2. marker 점스타일(값의 위치) ='문자'
#3. lw(line width)=숫자
#4. color = '문자' : 선과 마커의 색깔
#5. mfc(markerfacecolor) = '문자' 마커의 색깔
plt.show() #out 있는 상태는 출력상태가 아니다. show까지 해줘야 출력 상태
```

```python
import pandas as pd
```

```python
data=pd.read_csv('Traffic_Accident_2017.csv', encoding='euc-kr')
data
```

```python
#필요항목 : 요일 
data2=data['요일'].value_counts()
data2
```

```python

#요일 순서대로 가져오기
y=data2[['월','화','수','목','금','토','일']]
y
```

```python
#bar그래프 그리기
#x축 데이터 만들기
x=y.index
plt.bar(x,y)
plt.title('요일별 사고건수')
#x축, y축 제목 설정하기
plt.xlabel('요일')  # bar차트에서는 위에 label='x'안 써줘도 되네
plt.ylabel('사고건수').set_rotation(0)  # 축제목 회전하기
plt.ylim(480,620)
plt.show()
```

```python
#x=발생지시도(가나다순정렬)
#y=사상자수
#범위 '사고유형_대분류' == '차대차'
data
#data3=data[['사고유형_대분류','사상자수','발생지시도']]
data3=data[data['사고유형_대분류']=='차대차']
data3=data3[['발생지시도','사상자수']]
data3
```

```python
data3['발생지시도'].sort_values()
```

```python
#x=data3['발생지시도'].sort_values().value_counts().index 이걸 아래로 줄임
x=list(set(data3['발생지시도']))
x.sort()  # set쓰니까 정렬이 깨지네. 정렬은 언제나 나중에
x
```

```python
#y=data3[data3['발생지시도']==x[0]]['사상자수'].sum(axis=0)
y=[]
for i in x : 
    y.append(data3[data3['발생지시도']==i]['사상자수'].sum()) #axis=0는 기본값이니 생략가능
y
```

```python
plt.figure(figsize = (11,5))  # 그래프 사이즈. 젤 위에 써라 
plt.plot(x, y)	# line 그래프를 그립니다
plt.title('차vs차 교통사고의 사상자수')
#x축, y축 제목 설정하기
plt.xlabel('지역')  
plt.ylabel('사상자수')
plt.show()
```
